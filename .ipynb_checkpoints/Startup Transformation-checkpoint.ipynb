### import libraries

```python
from sklearn import preprocessing
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

### inspect and arrange data

```python
# load in financial data
financial_data = pd.read_csv('financial_data.csv')
```

```python
# print(financial_data.head())
month = financial_data["Month"]
revenue = financial_data["Revenue"]
expenses = financial_data["Expenses"]
```

```python
# plot of revenue over the past six months
plt.plot(month,revenue)
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Revenue')
plt.show()
```

```python
# plot of expenses over the past six months
plt.clf()
plt.plot(month,expenses)
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Expenses')
plt.show()
```

```python
# expenses data
expense_overview = pd.read_csv("expenses.csv")
# print(expense_overview.head(7))
# seperating expense data
expense_categories = expense_overview["Expense"]
proportions = expense_overview["Proportion"]
```

```python
# pie chart of expenses
plt.clf()
# creat other for a clearer chart
expense_categories = ['Salaries', 'Advertising', 'Office Rent', 'Other']
proportions = [0.62, 0.15, 0.15, 0.08]
plt.pie(proportions, labels = expense_categories,normalize=False)
plt.title("expenses pie chart")
plt.axis('Equal')
plt.tight_layout() 
plt.show()
```
